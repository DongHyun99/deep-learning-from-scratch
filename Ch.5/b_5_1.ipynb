# 5. 오차 역전파법

전 장에서 신경망의 가중치 매개변수의 기울기를 구할 때 수치미분을 이용해서 구했다.  
수치미분은 단순하고 구현이 쉽지만 계산시간이 매우 오래 걸린다는 단점이 있다.  
이러한 가중치 매개변수의 기울기를 효율적으로 계산하는 방법 중 **오차역전법**이 있다.  

# 5.1 계산 그래프  

계산 그래프는 계산 고자엉을 그래프로 나타낸 것이다.  
그래프는 Node와 edge를 가진 자료구조의 그래프이다.

_문제 2: 현빈 군은 슈퍼에서 사과를 2개, 귤을 3개 샀습니다. 사과는 1개에 100원, 귤은 1개에 150원입니다. 소비세가 10%일 때 지불금액을 구하세요._

<img src = "./image/fig_5_3.png" width="50%">

위와같이 계산 그래프는 다음 흐름으로 진행한다.  

    1. 계산 그래프를 구성한다.  
    2. 그래프에서 계산을 왼쪽에서 오른쪽으로 진행한다.  

2번처럼 계산을 왼쪽에서 오른쪽으로 진행하는 것을 **순전파(forward propagation)**라고 한다.  
반대로 오른쪽에서 왼쪽으로의 진행은 **역전파(backward propagation)**이라 한다.  
역전파는 이후 미분에서 중요한 역할을 할 것이다.  
 
## 5.1.2 국소적 계산  

계산 그래프의 특징: 국소적 계산(자신과 직접 관계된 작은 범위)을 전파함으로써 최종 결과를 얻음  

즉 전체적인 일에 신경쓸 필요없이 한 부분만 신경쓰면 된다.  
따라서 복잡한 계산을 국소적으로 계산하여 단순하게 여러번의 계산을 수행하도록 할 수 있음  

## 5.1.3 왜 계산 그래프로 푸는가? 

국소적 계산의 이점

    - 문제의 단순화  
    - 중간 계산 결과를 모두 보관이 가능  
    - 역전파를 통해 '미분'을 효율적으로 계산이 가능하다.  

<img src = "./image/fig_5_5.png" width="50%">  

위 그림과 같이 역전파는 1->1.1->2.2 이렇게 미분값을 전달 한다.  
즉 사과 1원이 오르면 최종 금액은 2.2원이 오른다는 뜻이다.  
중간 중간에 대한 미분 값도 국소적 계산을 통해 구할 수 있어 효율적이다.
