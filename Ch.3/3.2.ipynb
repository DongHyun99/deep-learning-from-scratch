## 3.2.3 계단 함수의 그래프

````python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.array(x>0, dtype=int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
````

## 3.2.4 시그모이드 함수 구현하기

````python
def sigmoid(x):
    return 1 / (1+np.exp(-x))

x=np.array([-1.0,1.0,2.0])
sigmoid(x)
````

````python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.show()
````

## 3.2.6 비선형 함수

이외 공통점: 두 함수 모두 비선형 함수  
신경망에서는 선형 함수를 사용하면 신경망의 층이 깊어져도 의미가 없어지기 때문에 비선형 함수를 이용해야 한다.  
왜냐하면 선형 함수는 은닉층이 없는 네트워크로도 쉽게 만들 수 있기 때문

## 3.2.7 ReLU 함수

```ReLU (렐루): Rectified Linear Unit```  
입력이 0이하이면 0을 출력하고 0이상이면 입력 값을 그대로 출력하는 함수  
수식으로 쓰면 h(x)는 x>0일 때 x를 출력, x<=0이면 0을 출력 하는 간단한 함수

````python
def relu(x):
    return np.maximum(0,x)
````
