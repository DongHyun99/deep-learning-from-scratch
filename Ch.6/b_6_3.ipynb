# 6.3 배치 정규화  

배치 정규화(batch Normalization): 각 층이 활성화를 적당히 퍼트리도록 '강제'하는 것  

## 6.3.1 배치 정규화 아록리즘  

배치 정규화가 주목 받는 이유  

    * 학습을 빨리 진행할 수 있다. (학습 속도의 개선)  
    * 초깃값에 크게 의존하지 않는다.  
    * 오버피팅을 억제한다. (드롭 아웃 등의 필요성 감소)  

배치 정규화의 기본 아이디어는 각 층의 활성화 값들이 적당히 분포되는 것  
따라서 아래와 같이 데이터 분포를 정규화 하는 배치 정규화 계층(Batch Norm Layer)을 신경망에 삽입한다.  

<img src = "./image/fig_6_16.png" width="40%">  

데이터의 분포가 평균이 0, 분산이 1이 되도록 정규화 한다.  

<img src = "./image/e_6_7.png">  

    - μ_B: m개의 입력 데이터 집합에 대한 평균  
    - σ_B^2: 분산  
    - ε: 매우 작은 값, 0으로 나누는 사태를 예방하는 역할  

이를 사용해서 데이터의 분포가 덜 치우치게 할 수 있다.  
또 배치 정규화 계층마다 이 정규화 된 데이터에 고유한 확대(scale)와 이동(shift)변환을 수행한다.  
수식은 다음과 같다.

ɣ가 확대, β가 이동을 담당한다.  초기값은 ɣ=1, β=0부터 시작하고, 학습하면서 적합한 값으로 조정해간다.

## 6.3.2 배치 정규화의 효과  

MNIST 데이터셋에 정규화 계층을 사용했을 때와 사용하지 않았을 때 학습진도가 어떻게 달라지는지를 살펴보자  

<img src = "./image/fig_6_18.png" width="40%">  

배치 정규화를 적용할 경우 학습을 빨리 진전 시킨다.  
초깃값 분초를 다양하게 줘가며 학습 진행이 어떻게 달라지는지 살펴보자  

<img src = "./image/fig_6_19.png" width="40%">  

거의 모든 경우에서 배치 정규화를사용할 때의 학습 진도가 빠른 것으로 나타난다.  
실제로 배치 정규화를 이용하지 않는 경우 초깃값이 잘 분포되어있지 않으면 학습이 전혀 진행되지 않는 경우도 있다.
